```python
# based on https://en.wikipedia.org/wiki/Bootstrap_aggregating
# graphs the percent of the sample in the training set (cardDeck) for various sample sizes (n_prime) 
# for sampling with replacement 
# uses monte carlo method to average across various samples
import matplotlib.pyplot as plt
from random import randint
from numpy import mean
n = 52
cardDeck = range(0, n)


# n_prime is sample size
n_prime = n

n_prime_values = list()
mean_coverages = list()

# iterate over possible sample sizes
for n_prime in range(1, 4 * n):
	coverageValues = list()
	numIterations = 20
	for i in range(0, numIterations):
		# sampling n_prime random items with replacement from the deck
		sample = set()
		for i in range(0, n_prime):
			randItem = randint(0, n)
			sample.add(randItem)

		# the percent of items in cardDeck also in the sample
		coverage = len(sample) / n
		coverageValues.append(coverage)

	meanCoverage = mean(coverageValues)
	n_prime_values.append(n_prime)
	mean_coverages.append(meanCoverage)

print("Plotting data...")
fig, ax = plt.subplots()  
plt.title("Sample Coverage for Sample Size (n') with Replacement n = 52")
ax.set_xlabel("sample size (n')")
ax.set_ylabel("% coverage")
ax.plot(n_prime_values, mean_coverages)
ax.vlines(x = n, ymin = 0, ymax = 1, label = "n")
ax.vlines(x = 2*n, ymin = 0, ymax = 1, label = "2n")
ax.vlines(x = 3*n, ymin = 0, ymax = 1, label = "3n")
plt.show()
```
